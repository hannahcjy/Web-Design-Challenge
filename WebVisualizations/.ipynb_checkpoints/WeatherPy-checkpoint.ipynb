```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
```

```python
# Convert array of JSONs into Pandas DataFrame
file = 'Resources\cities.csv'
city_data_pd = pd.read_csv(file)
```

```python
# Display the City Data Frame
city_data_pd = city_data_pd.drop(columns = "City_ID")
new = city_data_pd.rename(columns = {"Max Temp":"Max_Temp",
                                   "Wind Speed":"Wind_Speed"})
new.head()
```

## Latitude vs Temperature Plot

```python
plt.scatter(new["Lat"],new["Max_Temp"],
            edgecolor="darkblue", linewidths=1, marker="o", 
            alpha=0.4, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature" )
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Fig/Lat vs Temp.png")

# Show plot
plt.show()
```

## Latitude vs. Humidity Plot

```python
# Build the scatter plots for each city types
plt.scatter(new["Lat"],new["Humidity"],
            edgecolor="darkblue", linewidths=1, marker="o", 
            alpha=0.4, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Humidity")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Fig/Lat vs Humidity.png")

# Show plot
plt.show()
```

## Latitude vs. Cloudiness Plot

```python
# Build the scatter plots for each city types
plt.scatter(new["Lat"],new["Cloudiness"],
            edgecolor="darkblue", linewidths=1, marker="o", 
            alpha=0.4, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Cloudiness")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Fig/Lat vs Cloudiness.png")

# Show plot
plt.show()
```

## Latitude vs. Wind Speed Plot

```python
# Build the scatter plots for each city types
plt.scatter(new['Lat'], 
            new['Wind_Speed'],
            edgecolor="darkblue", linewidths=1, marker="o", 
            alpha=0.5, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Wind Speed")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Fig/Lat vs Wind speed.png")

# Show plot
plt.show()
```
